```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# separate the data
x=df[['YearsExperience']] #2d input
y=df['Salary'] #1d output
```

```python
# check linearity with scatter plot
plt.title('exp vs salary')
plt.xlabel('exp')
plt.ylabel('salary')
plt.scatter(x,y);
plt.grid();
```

```python
#build the model
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
yrexp=[[9.5]]
```

```python
lr.predict(yrexp)
```
