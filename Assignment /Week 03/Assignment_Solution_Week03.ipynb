# Section A — Classes and Objects

```python
"""1. Create a class BankAccount with the following:
a) Attributes: account_holder, account_number, and balance.
b) Methods: deposit(amount), withdraw(amount), and get_balance().
c) withdraw() should print a message and do nothing if the amount exceeds the current balance.
d) A __str__ method that returns a readable summary of the account.
Create two BankAccount objects and perform a few deposits and withdrawals on each""";
```

```python
class BankAccount:

    def __init__(self, account_holder, account_number, balance):
        self.account_holder = account_holder
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance + amount

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance = self.balance - amount

    def get_balance(self):
        return self.balance

    def __str__(self):
        return f"Name : {self.account_holder}\nAccount Number : {self.account_number}\nBalance : {self.balance}"


```

```python
acc1 = BankAccount("Shivam", 101, 50000)
acc2 = BankAccount("Rahul", 102, 30000)
```

```python
acc1.deposit(1000)
acc1.withdraw(2000)

acc2.deposit(500)
acc2.withdraw(4000)
```

```python
"""2. Create a class Student with attributes name, roll_no, and marks (a list of 5 subject marks). Add
methods:
a) calculate_percentage() — returns the average of all marks.
b) get_grade() — returns 'A' for >= 80, 'B' for >= 60, 'C' for >= 40, and 'Fail' otherwise.
c) display() — prints all details including percentage and grade.
Create a list of 3 Student objects and call display() on each.""";
```

```python
class Student:

    def __init__(self, name, roll_no, marks):
        self.name = name
        self.roll_no = roll_no
        self.marks = marks

    def calculate_percentage(self):
        total = sum(self.marks)
        percentage = total / 5
        return percentage

    def assign_grade(self):
        percentage = self.calculate_percentage()

        if percentage >= 75:
            return "A"
        elif percentage >= 60:
            return "B"
        elif percentage >= 40:
            return "C"
        else:
            return "Fail"

    def display_result(self):
        print("Name :", self.name)
        print("Roll No :", self.roll_no)
        print("Marks :", self.marks)
        print("Percentage :", self.calculate_percentage())
        print("Grade :", self.assign_grade())


```

```python
students = [
    Student("Shivam", 143, [80, 75, 70, 90, 85]),
    Student("Rahul", 144, [65, 70, 60, 75, 80]),
    Student("Amit", 145, [35, 40, 45, 50, 55])
]
```

```python
for s in students:
    s.display_result()
```

# Section B — Encapsulation

```python
"""3. Create a class Employee with private attributes __name, __emp_id, and __salary. Add:
a) Getter and setter methods for all three attributes.
b) The salary setter should raise a ValueError if the value is below 10000.
c) A method give_raise(percent) that increases salary by the given percentage.
Demonstrate that the private attributes cannot be accessed directly, and that all access goes
through the methods.""";
```

```python
class Employee:

    def __init__(self):
        self.__name = ""
        self.__emp_id = 0
        self.__salary = 0.0

    def set_data(self, name, emp_id, salary):
        if salary < 10000:
            raise ValueError("Salary cannot be below 10000")

        self.__name = name
        self.__emp_id = emp_id
        self.__salary = salary

    def get_data(self):
        print("Name :", self.__name)
        print("Emp ID :", self.__emp_id)
        print("Salary :", self.__salary)

    def give_raise(self, percent):
        self.__salary += self.__salary * percent / 100


```

```python
e = Employee()
```

```python

e.set_data("Shivam", 8, 67884.56)
e.give_raise(10)
e.get_data()
```

```python
"""4. Build a class SecureVault that stores a private __pin and a private __data dictionary. It should
have:
a) A method add_entry(pin, key, value) that adds to __data only if the pin matches.
b) A method get_entry(pin, key) that returns the value only if the pin matches.
c) Any wrong pin attempt should print 'Access Denied' and return nothing.""";
```

```python
class SecureVault:

    def __init__(self):
        self.__pin = 1234
        self.__data = {}

    def add_entry(self, pin, key, value):

        if pin == self.__pin:
            self.__data[key] = value
            print("Pin Verified")
        else:
            print("Access Denied")

    def get_entry(self, pin, key):

        if pin == self.__pin:
            print(self.__data.get(key))
        else:
            print("Access Denied")

```

```python

s = SecureVault()

s.add_entry(1234, 1, "ABC")

s.get_entry(1234, 1)
```

# Section C — Inheritance and Method Overriding

```python
"""5. Create a base class Shape with a method area() that returns 0 and a method describe() that
prints 'I am a shape'. Create three subclasses:
a) Circle(radius) — overrides area() using pi * r^2.
b) Rectangle(length, width) — overrides area().
c) Triangle(base, height) — overrides area().
Each subclass should also override describe() to print something specific about itself. Create one
object of each and call both methods.""";
```

```python
class Shape:
    def area(self):
        return 0

    def describe(self):
        print("I am a shape")


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius * self.radius

    def describe(self):
        print("I am a Circle")


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def describe(self):
        print("I am a Rectangle")


class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def area(self):
        return 0.5 * self.base * self.height

    def describe(self):
        print("I am a Triangle")


```

```python
# Creating Objects
c = Circle(5)
r = Rectangle(4, 6)
t = Triangle(8, 5)

# Calling Methods
c.describe()
print("Area =", c.area())

r.describe()
print("Area =", r.area())

t.describe()
print("Area =", t.area())
```

```python
"""6. Create a class Vehicle with attributes make, model, and year and a method start_engine() that
prints 'Engine started'. Create two subclasses:
a) Car(make, model, year, num_doors) — overrides start_engine() to include the car's name in
the message.
b) ElectricCar(make, model, year, battery_capacity) — overrides start_engine() to print 'Silent
start — battery at X kWh'.
Use super() in both subclasses to call the parent __init__. Create objects and test all methods.""";
```

```python
class Vehicle:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year

    def start_engine(self):
        print("Engine started")


class Car(Vehicle):
    def __init__(self, make, model, year, num_doors):
        super().__init__(make, model, year)
        self.num_doors = num_doors

    def start_engine(self):
        print("Engine started for", self.make, self.model)


class ElectricCar(Vehicle):
    def __init__(self, make, model, year, battery_capacity):
        super().__init__(make, model, year)
        self.battery_capacity = battery_capacity

    def start_engine(self):
        print("Silent start - battery at", self.battery_capacity, "kWh")


```

```python
# Creating Objects
v = Vehicle("Generic", "Vehicle", 2024)
c = Car("Maruti", "Swift", 2023, 4)
e = ElectricCar("Tesla", "Model 3", 2025, 75)

```

```python
# Testing Methods
v.start_engine()

c.start_engine()
print("Doors =", c.num_doors)

e.start_engine()
print("Battery Capacity =", e.battery_capacity, "kWh")
```

# Section D — Polymorphism

```python
"""7. Create a base class Animal with a method sound() that prints 'Some sound'. Create subclasses
Dog, Cat, and Cow, each overriding sound() with their respective sounds. Write a function
make_all_sounds(animals) that takes a list of Animal objects and calls sound() on each.
Demonstrate that the correct method is called regardless of the object type.""";
```

```python
class Animal:
    def sound(self):
        print("Some sound")


class Dog(Animal):
    def sound(self):
        print("Bark")


class Cat(Animal):
    def sound(self):
        print("Meow")


class Cow(Animal):
    def sound(self):
        print("Moo")


def make_all_sounds(animals):
    for animal in animals:
        animal.sound()


```

```python
# Creating objects
d = Dog()
c = Cat()
cw = Cow()

# List of Animal objects
animals = [d, c, cw]

# Calling function
make_all_sounds(animals)
```

```python
"""8. Create a class Calculator with overloaded-style methods using default arguments:
a) add(a, b, c=0) — adds two or three numbers.
b) multiply(a, b, c=1) — multiplies two or three numbers.
c) Also implement __add__ on a class Vector(x, y) so that two Vector objects can be added with
the + operator. Print the resulting vector.""";
```

```python
class Calculator:
    def add(self, a, b, c=0):
        return a + b + c

    def multiply(self, a, b, c=1):
        return a * b * c


class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def display(self):
        print("Vector(", self.x, ",", self.y, ")")
        
```

```python
# Calculator Object
c = Calculator()

print("Addition of 2 numbers =", c.add(10, 20))
print("Addition of 3 numbers =", c.add(10, 20, 30))

print("Multiplication of 2 numbers =", c.multiply(2, 3))
print("Multiplication of 3 numbers =", c.multiply(2, 3, 4))

# Vector Objects
v1 = Vector(2, 3)
v2 = Vector(4, 5)

v3 = v1 + v2

print("Resultant Vector:")
v3.display()
```

# Section F — Menu-Driven App with OOP

```python
"""10. Build a menu-driven Student Record Manager using OOP and file handling. The app should
have a class StudentManager with the following menu options:
1. Add Student — takes name, roll number, and marks for 3 subjects.
2. View All — displays all stored students with their average marks.
3. Search Student — finds a student by roll number.
4. Delete Student — removes a student by roll number.
5. Save to File — writes all records to students_data.txt.
6. Load from File — reads records back from students_data.txt on startup.
7. Exit
Requirements: use a class with a list as internal storage, handle invalid menu choices gracefully,
and use a while loop to keep the menu running until the user chooses to exit.""";
```

```python
class StudentManager:

    def __init__(self):
        self.students = []

    def add_student(self):
        name = input("Enter Name : ")
        roll = int(input("Enter Roll Number : "))

        m1 = int(input("Enter Subject1 Marks : "))
        m2 = int(input("Enter Subject2 Marks : "))
        m3 = int(input("Enter Subject3 Marks : "))

        self.students.append([name, roll, m1, m2, m3])

    def view_all(self):

        for i in self.students:

            avg = (i[2] + i[3] + i[4]) / 3

            print("Name =", i[0])
            print("Roll No =", i[1])
            print("Average =", avg)
            print()

    def search_student(self):

        roll = int(input("Enter Roll Number : "))

        for i in self.students:

            if i[1] == roll:
                print(i)
                return

        print("Student Not Found")

    def delete_student(self):

        roll = int(input("Enter Roll Number : "))

        for i in self.students:

            if i[1] == roll:
                self.students.remove(i)
                print("Record Deleted")
                return

        print("Student Not Found")

    def save_to_file(self):

        with open("students_data.txt", "w") as f:

            for i in self.students:

                f.write(str(i) + "\n")

        print("Data Saved")

   

obj = StudentManager()



while True:

    print("1.Add Student")
    print("2.View All")
    print("3.Search Student")
    print("4.Delete Student")
    print("5.Save To File")
    print("6.Exit")

    choice = int(input("Enter Choice : "))

    if choice == 1:
        obj.add_student()

    elif choice == 2:
        obj.view_all()

    elif choice == 3:
        obj.search_student()

    elif choice == 4:
        obj.delete_student()

    elif choice == 5:
        obj.save_to_file()

    elif choice == 6:
        break

    else:
        print("Invalid Choice")
```
