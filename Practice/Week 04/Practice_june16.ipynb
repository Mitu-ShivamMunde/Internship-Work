```python
import numpy as np
import time

# Create a list and an array with the same elements
list1 = [1, 2, 3, 4, 5]
arr = np.array([1, 2, 3, 4, 5])

# Calculate square of each element in the list
start = time.time()
result1 = [x**2 for x in list1]
list_time = time.time() - start

# Calculate square of each element in the array
start = time.time()
result2 = arr**2
array_time = time.time() - start

# Print results
print("List Result:", result1)
print("Array Result:", result2)

# Print processing times
print("List Time =", list_time)
print("Array Time =", array_time)

# Find how many times array is faster
speedup = list_time / array_time

print("NumPy array is", speedup, "times faster than list.")
```

```python
import numpy as np
import time

# Create a list and an array with the same elements
list1 = [1, 2, 3, 4, 5]
arr = np.array([1, 2, 3, 4, 5])

# Multiply each element by 5 in the list
start = time.time()
result1 = [x * 5 for x in list1]
list_time = time.time() - start

# Multiply each element by 5 in the array
start = time.time()
result2 = arr * 5
array_time = time.time() - start

# Print results
print("List Result:", result1)
print("Array Result:", result2)

print("List Time =", list_time)
print("Array Time =", array_time)

speedup = list_time / array_time
print("NumPy array is", speedup, "times faster than list.")
```
