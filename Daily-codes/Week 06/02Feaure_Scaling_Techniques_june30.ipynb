```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# check min and max values of Age and EstimatedSalary columns
```

```python
df.describe()
```

```python
# another way to check min and max values of Age and EstimatedSalary columns
print("Age column - Min:", df['Age'].min(), "Max:", df['Age'].max())
```

```python
print("EstimatedSalary column - Min:", df['EstimatedSalary'].min(), "Max:", df['EstimatedSalary'].max())
```

```python
plt.figure(figsize=(6,4))
plt.scatter(df['EstimatedSalary'],df['Age'], c = df['Purchased'])
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.title('Data Distribution')
plt.grid()
```

```python
#  separate the data
```

```python
x =df[['Age', 'EstimatedSalary']]
x
```

```python
# normalization
```

```python
from sklearn.preprocessing import MinMaxScaler  

sca = MinMaxScaler()
x_scaled = sca.fit_transform(x)
```

```python
x_scaled
```

```python
# check min and max values of each column 
```

```python
x_scaled[:, 0].min(), x_scaled[:, 0].max()  
```

```python
x_scaled[:,1].min(),x_scaled[:,1].max() 
```

```python
# plot the scatter chart to check the distribution 
```

```python
plt.figure(figsize=(6,4))
plt.scatter(x_scaled[:,1],x_scaled[:,0], c = df['Purchased'])
plt.xlabel('Age')
plt.ylabel('EstimatedSalary')
plt.title('Data Distribution')
plt.grid()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sca = StandardScaler()
x_scaled = sca.fit_transform(x)
x_scaled
```

```python
x_scaled[:, 0].min(), x_scaled[:, 0].max()
```

```python
x_scaled[:,1].min(),x_scaled[:,1].max()
```

```python
x_scaled.mean(axis=0)  # mean of each column
```

```python
x_scaled.std(axis=0)  # standard deviation of each column
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x_scaled[:, 0], x_scaled[:, 1],c = df['Purchased'])
plt.xlabel('Standardized Age')
plt.ylabel('Standardized EstimatedSalary')
plt.title('Age vs EstimatedSalary (Standardized)')
plt.grid()
```
