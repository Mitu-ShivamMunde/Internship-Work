## Outlier Detection and Removal

```python
x.dtypes
```

```python
x = pd.get_dummies(x,drop_first=True)
```

```python
col_order_ohe =x.columns.tolist()
```

```python
col_order_ohe
```

```python
# isolation forest for outlier removal
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso = IsolationForest(contamination= 0.05,random_state= 0)
```

```python
clean = iso.fit_predict(x)
```

```python
clean
```

```python
x = x[clean == 1]
y = y[clean == 1]
```

```python
x.shape
```

## Sampling (over sampling)

```python
y.shape
```

```python
y.value_counts()
```

```python
!pip install imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(k_neighbors=5)
```

```python
x_res,y_res = smote.fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
sns.countplot(x=y_res)
```

## Feature Selection

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skf = SelectKBest(k=5,score_func=chi2)
```

```python
x_new = skf.fit_transform(x_res,y_res)
```

```python
x_new.shape
```

```python
skf.get_support()
```

```python
skf.get_feature_names_out()
```

```python
x = x_res.loc[:,skf.get_support()]
```

```python
x;
```

```python
x.shape,y_res.shape
```

```python
x.describe
```

## Cross validation

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y_res, test_size=0.2, random_state=0,stratify=y_res
)
```

```python
x_train.shape,y_train.shape
```

## Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

sca = MinMaxScaler()

x_train_scaled = sca.fit_transform(x_train)
x_test_scaled = sca.transform(x_test)
```

```python
x_train_scaled.shape, y_train.shape
```

## Build the Model

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'random_state': [0, 42],
    'criterion': ['gini', 'entropy'],
    'max_features': ['sqrt', 'log2'],
    'n_estimators': [20, 30, 100]
}
```

```python
grid = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=params,
    scoring='accuracy',
    cv=5
)
```

```python
grid.fit(x_train_scaled,y_train)
```
