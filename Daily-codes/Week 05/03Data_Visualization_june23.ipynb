```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 23, 29, 34, 39, 45, 50])
y = np.array([14, 21, 31, 43, 48, 55, 58])
z = np.array([56, 50, 44, 37, 26, 20, 16])
```

```python
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("line chart: x vs y")
plt.plot(x,y, marker = 'o',mfc='red',ls='--');
plt.grid();
```

```python
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("line chart: x vs y")
plt.plot(x,y, marker = 'o',mfc='red',ls='--', label = '2024');
plt.plot(x,z, marker = 'o',mfc='black',ls='--', color = 'orange', label = '2025');
plt.grid();
plt.legend();
```

```python
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("line chart: x vs y")
plt.plot(x,y, marker = 'o',mfc='red',ls='--');
plt.grid();

plt.subplot(2,2,2)
plt.xlabel("x value")
plt.ylabel("z value")
plt.title("Line chart: x vs z")
plt.plot(x, z, marker='*', mfc='green', ls='--');
plt.grid();
```

# Bar chart

```python
courses = ['python', 'lisp', 'c', 'R', 'php']
students = [68, 30, 47, 39, 60]
```

```python
plt.bar(courses, students);
plt.xlabel("courses value")
plt.ylabel("student value")
plt.title("line chart: courses value vs student value ")
```

```python
plt.title("student enrollment")
plt.xlabel("course")
plt.ylabel("no.of student enrolled")
plt.bar(courses,students,color ='purple',edgecolor ='red');
plt.grid();
```

```python
# load ->Social_Network_Ads.csv 
# plot bar graph showing count of how many pepole purchased and how many did not 
# Purchased column -> 0 = purchased, 1 = not purchased
```

```python



ads = pd.read_csv("Social_Network_Ads.csv")
ads
```

```python
count = ads['Purchased'].value_counts()
```

```python
 ads['Purchased'].value_counts().index
```

```python
ads['Purchased'].value_counts().values
```

```python

plt.title("Purchase Count")
plt.xlabel("Status")
plt.ylabel("Number of People")

plt.bar(
    ads['Purchased'].value_counts().index,
    ads['Purchased'].value_counts().values,
    color=['green', 'red']
)

plt.grid()
plt.show()
```
