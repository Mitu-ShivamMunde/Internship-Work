# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([10, 23, 29, 34, 39, 45, 50])
y = np.array([14, 21, 31, 43, 48, 55, 58])
z = np.array([56, 50, 44, 37, 26, 20, 16])
```

# line chart

```python
plt.plot(x,y);
```

```python
plt.xlabel("x value")
plt.ylabel("y value")
plt.title("line chart: x vs y")
plt.plot(x,y, marker = 'o',mfc='red',ls='--');
plt.grid();
# o,*,H,<,>,^
```

```python
plt.xlabel("x value")
plt.ylabel("z value")
plt.title("Line chart: x vs z")
plt.plot(x, z, marker='*', mfc='green', ls='--')
plt.grid()
```
