```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x =[12,19,23,29,33,37,41,44,48,57]
y =[15,21,25,31,37,40,43,49,53,61]
z =[63,59,55,51,46,39,35,28,22,10]
```

# line plot

```python

plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,y,marker ='o',mfc ='red',ls ="--");
plt.grid()
# markers-> <,>,^,*,H,D,o
```

```python
plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,z,marker ='o',mfc ='red',ls ="--");
plt.grid()
# markers-> <,>,^,*,H,D,o
```

```python
plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,y,marker ='o',mfc ='red',ls ="--",color ='purple');
plt.plot(x,z,marker ='o',mfc ='red',ls ="--",color ='black');
plt.grid()
# markers-> <,>,^,*,H,D,o
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,y,marker ='o',mfc ='red',ls ="--");
plt.grid()
# markers-> <,>,^,*,H,D,o
plt.subplot(2,2,2)
plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,z,marker ='o',mfc ='red',ls ="--");
plt.grid()
# markers-> <,>,^,*,H,D,o

```

# Bar plot

```python
courses =['python','kotlin','go','swift','dart']
students =[60,56,40,45,63]
```

```python
plt.title('student enrollment')
plt.xlabel('courses')
plt.ylabel('no of students')
plt.bar(courses,students,color = 'black',width=0.3)
plt.grid();
```

```python
df = pd.read_csv('student.csv')
df.head(3);
```

```python
plt.title('student enrollment')
plt.xlabel('names')
plt.ylabel('marks')
plt.xticks(rotation=45)
plt.bar(df['name'],df['marks'],color='black',width=0.3)
plt.grid();
```

# pie

```python
plt.pie(students, labels = courses,autopct="%.2f%%",
        shadow = True, explode = [0,0.2,0,0,0]);
```

# scatter plot

```python
plt.scatter(x, y, color ='black',marker ='o');
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title('student enrollment')
plt.xlabel('names')
plt.ylabel('marks')
plt.xticks(rotation=45)
plt.bar(df['name'],df['marks'],color='black',width=0.3)
plt.grid();



plt.subplot(2,2,2)
plt.title('student enrollment')
plt.xlabel('courses')
plt.ylabel('no of students')
plt.bar(courses,students,color = 'black',width=0.3)
plt.grid();


plt.subplot(2,2,3)
plt.title('line graph')
plt.xlabel('x values')
plt.xlabel(' z values')
plt.plot(x,y,marker ='o',mfc ='red',ls ="--");
plt.grid()

plt.subplot(2,2,4)
plt.pie(students, labels = courses,autopct="%.2f%%",
        shadow = True, explode = [0,0.2,0,0,0]);

plt.savefig("mypolt.png")

```
